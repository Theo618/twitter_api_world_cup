```python
import json
```

```python
Data = []
for doc in os.listdir():
    if "raw" in doc :
        with open(doc, 'r', encoding="utf-8") as file:
            for line in file :
                Data.append(json.loads(line))
```

```python
var = []
for elt in Data : 
    l=list(elt.keys())
    var.extend(l)
```
